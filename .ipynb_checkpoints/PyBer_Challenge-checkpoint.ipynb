### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
city_data_df
```

```python
ride_data_df
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
# use the groupby() function to create a Series of data that has the type of city as the index, 
# then apply the count() method to the "ride_id" column.
total_ride_by_type_of_city = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_ride_by_type_of_city
```

```python
# 2. Get the total drivers for each city type
total_driver_by_type_of_city = city_data_df.groupby(["type"]).sum()["driver_count"] 
total_driver_by_type_of_city
```

```python
#  3. Get the total amount of fares for each city type
# use the groupby() function to create a Series of data that has the type of city 
# as the index, then apply the sum() method to the "fare" column.
total_amount_of_fares_by_city_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_amount_of_fares_by_city_type
```

```python
#  4. Get the average fare per ride for each city type. 
# calculate the average fare per ride by city type by dividing the sum of all the fares by the total rides.
avg_per_ride = total_amount_of_fares_by_city_type / total_ride_by_type_of_city  
avg_per_ride
```

```python
# 5. Get the average fare per driver for each city type. 
# calculate the average fare per driver by city type by dividing the sum of all the fares by the total drivers.
avg_fare_per_driver = total_amount_of_fares_by_city_type / total_driver_by_type_of_city
avg_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame. with all the data gathered from Steps 1-5
# total rides = number of rides
# total drivers = sum of drivers
# total fare = sum of total fare for each city type
#
pyber_summary_df = pd.DataFrame(
          {"Total Rides": total_ride_by_type_of_city, 
          "Total Drivers": total_driver_by_type_of_city, 
          "Total Fares": total_amount_of_fares_by_city_type,
           "Average Fare per Ride": avg_per_ride, 
           "Average Fare per Driver": avg_fare_per_driver})
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Rides"] = pyber_summary_df["Total Rides"].map("{:,}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("{:.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("{:.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("{:.2f}".format)
pyber_summary_df
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
total_fare_amount_for_each_date = pyber_data_df.groupby(["type","date"]).sum()["fare"]
total_fare_amount_for_each_date
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
total_fare_amount_for_each_date = total_fare_amount_for_each_date.reset_index()
total_fare_amount_for_each_date
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pivot_date_df=total_fare_amount_for_each_date.pivot('date', 'type', 'fare')
# pivot_date_df = total_fare_amount_for_each_date.pivot('date', 'type', 'fare').loc['2019-01-01':'2019-04-29']
pivot_date_df
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pivot_date_df = pivot_date_df.loc['2019-01-01':'2019-04-29']
pivot_date_df
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
pivot_date_df.index=pd.to_datetime(pivot_date_df.index)
print('pivot datetime as index: ',pivot_date_df.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pivot_date_df.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
resampled_df = pivot_date_df.resample('W', origin='2019-01-06').sum()
resampled_df
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
# in Step 8, graph the resampled DataFrame from Step 7 using the object-oriented interface 
# method and the df.plot() method, as well as the Matplotlib "fivethirtyeight" graph style code 
# snippet provided in the starter code. Annotate the y-axis label and the title, 
# then use the appropriate code to save the figure as PyBer_fare_summary.png in your "analysis" folder.
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

df = pd.DataFrame(resampled_df)
#pass the title, ylabel and figure size to matplotlib
df.plot(ylabel="Fare($USD)", title="Total Fare by City Type",figsize = (20,6))
plt.tight_layout()
plt.savefig("analysis/PyBer_fare_summary.png")
```
